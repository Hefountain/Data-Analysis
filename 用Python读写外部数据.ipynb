```python
# 创建csv文件
with open('./csv1.csv','a') as fw:
    writer = csv.writer(fw)
    writer.writerow(['c1', 'c2', 'c3'])
    for x in range(10):
        writer.writerow([x, chr(ord('a') + x), 'abc'])
        
# ord()函数用于返回一个字符的ASCII值，如 ord('a') 返回97
# chr()函数正好相反，返回ASCII值所对应的字符 

# csv.writer() 不会自动为字符串加上双引号，如需要，则：
# csv.writer(fw, quoting=csv.QUOTE_NONNUMERIC)   ，csv.QUOTE_NONNUMERIC为所有的非数字类型增加双引号
```

```python
with open('./csv1.csv','a') as fw:
    writer = csv.writer(fw, quoting=csv.QUOTE_NONNUMERIC)
#     writer.writerow(['c1', 'c2','c3'])
    for x in range(10):
        writer.writerow([x, chr(ord('a') + x), 'abc'])
```

```python
#  将读取的结果转化为字典    csv.DictReader()
with open('./csv1.csv','r') as fr:
    rows = csv.DictReader(fr)
    for row in rows:
        print (row)
```

```python
from __future__ import print_function
import pandas as pd
from pandas import read_excel
# pd.set_option('display.max_columns', 5)
# pd.set_option('display.max_rows', 10)
df = read_excel('/Users/fountain/Downloads/data_science_tool_book_code-master/chapter7/A0202.xls', 'Sheet1')
print(df)
```

```python
# read_excel()函数中的参数   index_col和 skiprows 分别表示：
# 告诉程序哪一行是序号行，可以隐藏
# 第二个参数表示可以跳过Excel开头的几行，不去读取这些行的信息
df = read_excel('/Users/fountain/Downloads/data_science_tool_book_code-master/chapter7/A0202.xls', 'Sheet1',
               index_col=0, skiprows=3)
print (df)
```

```python
# 也可以用上下文管理器with打开excel文件, 如果一个excel有多个Sheet
with pd.ExcelFile('/Users/fountain/Downloads/data_science_tool_book_code-master/chapter7/A0202.xls') as xls:
    for x in range(1,2):
        df = read_excel(xls, 'Sheet{}'.format(x), index_col=0, skiprows=3)
        print(df)  
        
# 理解： 'Sheet{}'.format(x)   ——> format函数格式化字符串
```

```python
from __future__ import print_function
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 180)
pd.set_option('max_colwidth', 110)

# 探索性别对幸存率的影响
data = pd.read_csv('/Users/fountain/Downloads/data_science_tool_book_code-master/chapter10/train.csv')
print(data)
print(data.xs(0))

# 计算男性和女性的幸存率
# 方法1
men = data[data.Sex == 'male']
women = data[data.Sex == 'female']
proportion_women_survived = float(sum(women.Survived)) / len(women)
proportion_men_survived = float(sum(men.Survived)) / len(men)
print('female: ', proportion_women_survived)
print('male: ', proportion_men_survived)

# 方法2
print(data.groupby('Sex').Survived.mean())

# 探索年龄对幸存率的影响
import matplotlib.style

matplotlib.style.use('ggplot')

need_data = data.loc[:, ['Age', 'Survived']].dropna(how='any')
need_data['Age'] = need_data['Age'].apply(round).astype('int16')
grouped = need_data.groupby('Age').Survived
survived = grouped.sum()
died = grouped.size() - survived
df = pd.DataFrame(dict(died=died, survived=survived))
df.plot.bar(figsize=(20, 10), color=('k', '#5EB2BF'))

plt.show()

# 探索阶级对幸存率的影响

if __name__ == '__main__':
    pass
```
