```python
#  柱状图
loadata = pd.DataFrame(pd.read_csv('/Users/fountain/Downloads/data_science_tool_book_code-master/chapter7/test.csv'))
loa_grade = loadata.groupby('Embarked')['Fare'].agg(sum)
loa_grade
```

```python
import matplotlib.patches as mpatches
a=np.array([1,2,3])
plt.bar([1,2,3], loa_grade,color='#99CC01', alpha=0.8, align='center', edgecolor='white',width=0.4)
plt.xlabel(u'乘客等级',fontproperties=font_set)
plt.ylabel(u'船费',fontproperties=font_set)
plt.title(u'不同等级乘客的船费分布',fontproperties=font_set)

red_patch = mpatches.Patch(color='k', label=u'乘客船费')
plt.legend(handles=[red_patch],prop=font_set,fontsize='xx-small')

# plt.legend(['乘客船费'],loc='upper left')
plt.grid(color='#95a5a6', linestyle='--', linewidth=1, axis='y', alpha=0.5)
plt.xticks(a,(u'C级',u'Q级',u's级'),fontproperties=font_set)
#显示图表
plt.show()
```
