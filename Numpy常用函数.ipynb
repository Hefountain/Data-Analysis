```python
import numpy as np
# usecols 的参数为一个元组
c,v = np.loadtxt('/Users/fountain/Desktop/data.csv',delimiter=',', usecols=(6,7), unpack=True)
# 加权平均值是由每个分量乘以反映其重要性的因子得到的平均值。
#考虑数组[1,2,3,4]和相应的权重[4,3,2,1]，通过将相应元素的乘积相加，并将和除以权重的和，来计算加权平均值。
vwap = np.average(c,weights=v)   #  计算成交量加权平均价格
print "VWAP=",vwap
```

```python
t = np.arange(len(c))
twap = np.average(c,weights=t)
print "TWAP=",twap    #计算时间加权平均价格
```

```python
# usecols的参数为元组，它的元素个数为1个，为避免歧义，所以在后面多加了一个逗号表示确是一个元组，而不是其它 
c = np.loadtxt("/Users/fountain/Desktop/data.csv", delimiter=",", usecols=(6,),unpack=True)
print "中位数为",np.median(c)
```

```python
# 对数组进行排序   msort()函数
sorted_close = np.msort(c)
sorted_close
```

```python
# 上面的数组为个数为偶数，所以中位数应该等于中间那两个数的平均值
N = len(c)
aver = (sorted_close[N/2] + sorted_close[(N-1)/2])/2
print "中位数为",aver
```

```python
# 计算方差：是指各个数据与所有数据算术平均数的离差平方和除以数据个数所得到的值
print "方差为",np.var(c)
```

```python
# 根据方差的定义即：
variance = np.mean((c-c.mean())**2)   # ndarray对象也有mean方法，所以数组c可以直接调用mean方法
print "方差为",variance
```

```python
#  注意样本方差和总体方差的区别。总体方差是用数据个数去除离差平方和（离差:指单项数值和平均值的差），而样本方
# 差则是用样本数据个数减1去除离差平方和，其中样本数据个数减1（即n-1）称为自由度。之所以有这样的差别，是为了
# 保证样本方差是一个无偏估计量。
```

```python
import numpy as np
c = np.loadtxt("/Users/fountain/Desktop/data.csv",delimiter=",", usecols=(6,),unpack=True)
returns = np.diff(c)/c[:-1]
print returns
print "标准差=",np.std(returns)
logreturns = np.diff(np.log(c))   #  计算对数收益率
pos = np.where(returns > 0)   #  返回正数对应的索引
print pos  
annual_volatility = np.std(logreturns)/np.mean(logreturns)
annual_volatility = annual_volatility / np.sqrt(1./252.)  #  计算年波动率
print "年波动率=",annual_volatility
month_volatility = annual_volatility*np.sqrt(1./12.)
print "月波动率=",month_volatility
```
